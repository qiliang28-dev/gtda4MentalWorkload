```python
import pandas as pd
import os
import copy
import numpy as np
```

### 查看结果的方法

   1. 解注释下方的cross_val变量
   2. 然后依次运行下面代码，即可看到最后结果
   3. shift+回车键 快捷执行代码

```python
model = "LR" #  [RF, LR, KNN, MLP]四选一

cross_val = "intra_sub_rs"
#cross_val = "LOSO_rs"

base_dir = f'./all_result/{model}'
cur_dir = os.path.join(base_dir, cross_val)
rs_dir = os.listdir(cur_dir)
rs_dir_copy = copy.deepcopy(rs_dir)
for rid, rd in enumerate(rs_dir_copy):
    if '.' in rd:
        rs_dir.pop(rid)
print(rs_dir)
```

```python
PW0_mean = np.zeros((len(rs_dir), 4), dtype=float)
PW0_std = np.zeros((len(rs_dir), 4), dtype=float)
PW1_mean = np.zeros((len(rs_dir), 4), dtype=float)
PW1_std = np.zeros((len(rs_dir), 4), dtype=float)
PW2_mean = np.zeros((len(rs_dir), 4), dtype=float)
PW2_std = np.zeros((len(rs_dir), 4), dtype=float)
for rsd in rs_dir:
    PW0_mean[int(rsd)-1, 0] = int(rsd)
    PW0_std[int(rsd)-1, 0] = int(rsd)
    PW1_mean[int(rsd)-1, 0] = int(rsd)
    PW1_std[int(rsd)-1, 0] = int(rsd)
    PW2_mean[int(rsd)-1, 0] = int(rsd)
    PW2_std[int(rsd)-1, 0] = int(rsd)
    
    cur_rs_dir = os.path.join(cur_dir, rsd)
    all_rs = os.listdir(cur_rs_dir)
    PW_0_mean = []
    PW_0_std = []
    PW_1_mean = []
    PW_1_std = []
    PW_2_mean = []
    PW_2_std = []
    
    for cur_file in all_rs:
        if not str(cur_file).endswith(".csv"):
            continue
        cur_file_dir = os.path.join(cur_rs_dir, cur_file)
        pw_i = int(str(cur_file).split(".")[0][-1])
        if cross_val != "LOSO_rs":
            # read intra-sub result out
            rs = np.loadtxt(cur_file_dir, dtype=float, delimiter=",", skiprows=0, usecols=[0, 2, 4])
            # rs = pd.read_csv(cur_file_dir, sep=",", names=["AUC_ROC", "Kappa", "Acc", "macro_P", "weight_P", "f1"])
        else:
            # read loso result out
            rs = np.loadtxt(cur_file_dir, dtype=float, delimiter=",", skiprows=1, usecols=[0, 2, 4])
            # rs = pd.read_csv(cur_file_dir, sep=",", names=["AUC_ROC", "Kappa", "Acc", "macro_P", "weight_P", "f1"], header=0)
            
        mean = np.mean(rs, axis=0, keepdims=True)
        std = np.std(rs, axis=0, keepdims=True)
        # print(std)
        #print(mean.shape, std.shape)
        if pw_i == 0:
            PW_0_mean.append(mean)
            PW_0_std.append(std)
        elif pw_i == 1:
            PW_1_mean.append(mean)
            PW_1_std.append(std)
        elif pw_i == 2:
            PW_2_mean.append(mean)
            PW_2_std.append(std)
        else:
            print(pw_i)
    for pw_level in range(3): 
        print("\n")
        print(f"PW等级为：{pw_level}, 当前窗口为：{rsd}s")    
        print(pd.DataFrame(np.concatenate(eval(str(f"PW_{pw_level}_mean")), axis=0), columns=['AUC_ROC', 'Acc', 'F1-score'], dtype=float))
        
    PW0_mean[int(rsd)-1, 1:] = np.mean(np.concatenate(PW_0_mean, axis=0), axis=0, keepdims=True)[0]
    PW0_std[int(rsd)-1, 1:] = np.std(np.concatenate(PW_0_std, axis=0), axis=0, keepdims=True)[0] if cross_val != "LOSO_rs" else np.concatenate(PW_0_std, axis=0)
    PW1_mean[int(rsd)-1, 1:] = np.mean(np.concatenate(PW_1_mean, axis=0), axis=0, keepdims=True)[0]
    PW1_std[int(rsd)-1, 1:] = np.std(np.concatenate(PW_1_std, axis=0), axis=0, keepdims=True)[0] if cross_val != "LOSO_rs" else np.concatenate(PW_1_std, axis=0)
    PW2_mean[int(rsd)-1, 1:] = np.mean(np.concatenate(PW_2_mean, axis=0), axis=0, keepdims=True)[0]
    PW2_std[int(rsd)-1, 1:] = np.std(np.concatenate(PW_2_std, axis=0), axis=0, keepdims=True)[0] if cross_val != "LOSO_rs" else np.concatenate(PW_2_std, axis=0)
```

```python
rs = [PW0_mean,PW1_mean, PW2_mean, PW0_std, PW1_std, PW2_std]
rs_name = ["PW0_mean","PW1_mean", "PW2_mean", "PW0_std", "PW1_std", "PW2_std"]
for name, content in zip(rs_name, rs):
    print(f"--------------------------{name}-----------------------------")
    c = pd.DataFrame(content, columns=['SW', 'AUC_ROC', 'Acc', 'F1-score'], dtype=float)
    print(c)
    print("\n")
```
